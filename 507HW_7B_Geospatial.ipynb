```python
import geopandas as gpd
```

### Dataset 1

```python
df_gpd = gpd.read_file('National_Obesity_By_State.geojson')
df_gpd [:10]
```

##### Dropping rows with any missing values

```python
df_gpd.dropna(inplace=True)
df_gpd
```

##### Map Visualization

```python
df_gpd.plot("Obesity", legend=True, colormap="viridis")
```

##### This dataset depicts obesity by state across the United States. This visualization portrays the obesity rates of the country. The attributes showcases the shape length and area of the state which creates an accurate visualization of the map, and the obesity rates are shown with lower obesity rates in darker colors and higher rates in lighter colors. The color I choose was in "viridis", which assigns different colors to different values in the "obesity" column. If I chose just one color, then the whole map would be of that color and we wouldn't be able to visualize the data.

### Dataset 2

```python
df_gpd2 = gpd.read_file('Grocery_Stores.geojson')
df_gpd2 [:10]
```

##### Dropping rows with any missing values

```python
df_gpd2.dropna(inplace=True)
df_gpd2
```

##### Map Visualization

```python
df_gpd2.plot("ZipCode", legend=True, color="red")
```

This dataset depicts grocery store locations in the city of Detroit. This visualization portrays the grocery zip codes of the state. The attributes consists of the company of grocery store, common name, and contact info. I wanted to showcase the grocery store locations across the state, and we can see that some locations are more sparse while some are more dense with stores. The color I choose was red-I went with a single color because I didn't have the differentiate between the different stores given the zip code location. If I chose a variety of colors, the map would have each dot be a different color and it would have blended together when the shades are very similar.
